## 2.5 Automatic Differentiation
PyTorch can only automatically start the "backward" chain-rule process from a single number (a scalar)

Let’s assume that we are interested in differentiating the function y = 2 * x_t * x with respect to the column vector x
. To start, we assign x an initial value.

```python
x = torch.arange(4.0)
x
```

```python
x.requires_grad_(True)
x.grad # The gradient is None by default
```

```python
y = 2 * torch.dot(x, x)
y
```

```python
y.backward()
x.grad
```

```python
x.grad == 4 * x
```

```python
x.grad.zero_()
y = x * x
y.backward(gradient=torch.ones(len(y)))  # Faster: y.sum().backward()
x.grad
```

```python
# Detach computation
x.grad.zero_()
y = x * x
u = y.detach()
z = u * x

z.sum().backward()
x.grad == u
```
